```python
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
from datetime import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.colors as colors
%matplotlib inline

# 99.9th percentile
# File path to the CSV
csv_file = "/scratch/k10/ef7927/research_project/csv/all_stations/all_stations.csv"

# Load the CSV file
df = pd.read_csv(csv_file)

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Round the datetime to the nearest hour
df['rounded_hour'] = df['date'].dt.round('h')

# Extract the hour from the rounded datetime
df['hour'] = df['rounded_hour'].dt.hour

# Count the number of extreme precipitation events (rr) for each hour across 24 hours
hourly_event_count = df.groupby('hour')['rr'].count()

# Display the event count for each hour
print(hourly_event_count)

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(hourly_event_count.index, hourly_event_count.values, color='skyblue')

# Set the x and y axis labels with font size
plt.xlabel('Hour (UTC)', fontsize=14)
plt.ylabel('Number of Extreme Events', fontsize=14)

# Set title font size
#plt.title('Hourly Count of Extreme Events (2015-2023)', fontsize=20)

# Adjust x-tick and y-tick label sizes
plt.xticks(range(24), fontsize=14)  # Set x-tick size to 12
plt.yticks(fontsize=14)  # Set y-tick size to 12

# Add a grid to the y-axis
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Uncomment if you want a threshold line
#plt.axhline(y=400, color='red', linestyle='--', linewidth=2, label='Threshold = 400')

# Display the plot
plt.tight_layout()
plt.show()

# If you want to save the results to a CSV file
output_file = "/scratch/k10/ef7927/research_project/csv/all_stations/999_extreme_event_hourly_counts.csv"
hourly_event_count.to_csv(output_file)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from Excel file
data = pd.read_csv('/scratch/k10/ef7927/research_project/codes/general/rr_average.csv')

# Extract months, places, rainfall, and event numbers
months = data.iloc[:, 0].tolist()
places = data.columns[1:-1].tolist()  # Exclude the last column (which is for event numbers)
event_numbers = data.iloc[:, -1].tolist()

# Creating the figure
fig, ax1 = plt.subplots(figsize=(10, 6))

# Iterate over each place
for place in places:
    # Get data for the current place
    rainfall = data[place].tolist()

    # Plotting the first data (rainfall) on the primary y-axis
    ax1.plot(months, rainfall, label='Monthly Average')

# Creating the secondary y-axis for event numbers
ax2 = ax1.twinx()

# Plotting the second data (event numbers) on the secondary y-axis
ax2.plot(months, event_numbers, color='red', linestyle='--', label='Number of Events')

# Set labels and title
ax1.set_xlabel('Month', fontsize=14)
ax1.set_ylabel('Monthly Average of Precipitation (mm)', fontsize=14)
ax2.set_ylabel('Number of Events', fontsize=14, rotation=270, labelpad=15)
#plt.title('Distribution of Monthly Average of Precipitation and Number of Extreme Precipitation\nof 20 Stations from 2015 to 2023', fontsize=10)

# Set x-axis tick size
ax1.tick_params(axis='x', labelsize=14)  # Change 12 to your desired font size

# Adding legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines1 + lines2, labels1 + labels2, loc='upper center', fontsize='12')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, fontsize=20)

# Show plot
fig = plt.gcf()
fig.tight_layout(rect=[0.5, 0.5, 1.5, 1.4])
plt.show()
```
