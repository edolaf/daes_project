# Packages

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import box
from matplotlib.colors import ListedColormap, BoundaryNorm
import pandas as pd
import matplotlib.patches as patches
from scipy.ndimage import binary_dilation
```

# Functions and constants

```python
# Define map bounds for Indo (full region)
lat_indo = [-14.0, 8.0]
lon_indo = [92, 141]
lat_min_indo, lat_max_indo = min(lat_indo), max(lat_indo)
lon_min_indo, lon_max_indo = min(lon_indo), max(lon_indo)
bbox_indo = box(lon_min_indo, lat_min_indo, lon_max_indo, lat_max_indo)

# Define map bounds for West Sumatra (zoom-in region)
lat_west = [-3.5, 1.2]
lon_west = [97, 103]
lat_min_west, lat_max_west = min(lat_west), max(lat_west)
lon_min_west, lon_max_west = min(lon_west), max(lon_west)
bbox_west = box(lon_min_west, lat_min_west, lon_max_west, lat_max_west)

# Define elevation bins and colors (consistent configuration)
elevation_bins = [0, 300, 600, 900, 1200, 1500, 1800, 2100]
elevation_colors = [
    "#1a9850", "#66bd63", "#a6d96a", "#d9ef8b", "#fee08b",
    "#fdae61", "#f46d43", "#d73027",
]

custom_cmap = ListedColormap(elevation_colors)
norm = BoundaryNorm(elevation_bins, custom_cmap.N)
sea_color = np.array([135/255, 206/255, 235/255, 1])  # Light blue for the ocean
```

# Read data

## GeoTiff file

```python
tiff_file = 'output_GEBCOSubIceTopo_Indo.tif'

```

## Spreadsheet

```python
locations = pd.read_csv('station_coordinate.csv', index_col=0)
locations
```

## SHP File/Coastline

```python
coastline_file ='prov.shp'
coastline = gpd.read_file(coastline_file).clip(bbox_indo)
coastline_west = gpd.read_file(coastline_file).clip(bbox_west)
```

# Analysis

```python
with rasterio.open(tiff_file) as src:
    elevation_data = src.read(1).astype(float)
    elevation_data[elevation_data == -9999] = np.nan

    # **Land and Sea Masking**
    land_mask = elevation_data > 0  # Identify land pixels
    sea_mask = elevation_data <= 0  # Identify sea pixels

    # **Fix Inland Water Bodies**
    dilated_land = binary_dilation(land_mask, iterations=2)  # Expand land areas
    inland_sea = np.logical_and(dilated_land, sea_mask)  # Detect inland water
    elevation_data[inland_sea] = 10  # Set inland sea pixels to 10m elevation

    # Update land-only data
    elevation_land_only = np.where(elevation_data < 0, np.nan, elevation_data)

    # Extract Indo region
    row_min_indo, col_min_indo = src.index(lon_min_indo, lat_max_indo)
    row_max_indo, col_max_indo = src.index(lon_max_indo, lat_min_indo)

    # Ensure correct slicing order
    if row_min_indo > row_max_indo:
        row_min_indo, row_max_indo = row_max_indo, row_min_indo
    if col_min_indo > col_max_indo:
        col_min_indo, col_max_indo = col_max_indo, col_min_indo

    elevation_subset_indo = elevation_data[row_min_indo:row_max_indo, col_min_indo:col_max_indo]
    elevation_land_subset_indo = np.where(elevation_subset_indo < 0, np.nan, elevation_subset_indo)

    # Extract West Sumatra region
    row_min_west, col_min_west = src.index(lon_min_west, lat_max_west)
    row_max_west, col_max_west = src.index(lon_max_west, lat_min_west)

    elevation_subset_west = elevation_data[row_min_west:row_max_west, col_min_west:col_max_west]
    elevation_land_subset_west = elevation_land_only[row_min_west:row_max_west, col_min_west:col_max_west]
```

# Plot

```python
# Create subplot (upper and lower panels)
fig = plt.figure(figsize=(12, 10))  # Adjust figure size
gs = fig.add_gridspec(2, 1, height_ratios=[1.75, 1])  # Adjust height ratio

# **Upper Panel (West Sumatra - Larger)**
ax_west = fig.add_subplot(gs[0])
ax_west.imshow(elevation_subset_west < 0, extent=[lon_min_west, lon_max_west, lat_min_west, lat_max_west], cmap=ListedColormap([sea_color]))
img_west = ax_west.imshow(elevation_land_subset_west, extent=[lon_min_west, lon_max_west, lat_min_west, lat_max_west], cmap=custom_cmap, norm=norm, alpha=0.9)
coastline_west.boundary.plot(ax=ax_west, color='black', linewidth=1)
ax_west.scatter(locations['longitude'], locations['latitude'], color='blue', marker='o', edgecolors='w', linewidths=0.7, s=30, label='Stations', zorder=3)
#ax_west.set_title('Topography Map of West Sumatra', fontsize=10)
ax_west.set_xlabel('Longitude', fontsize=12)
ax_west.set_ylabel('Latitude', fontsize=12)
ax_west.set_xticks(np.arange(97, 103, 1))
ax_west.set_yticks(np.arange(-3, 2, 1))
ax_west.grid(True, linestyle="--", linewidth=0.7, alpha=0.7)
ax_west.legend(fontsize=12)

# **Lower Panel (Indonesia - Same Width, Adjusted Height)**
ax_indo = fig.add_subplot(gs[1])
ax_indo.imshow(elevation_subset_indo < 0, extent=[lon_min_indo, lon_max_indo, lat_min_indo, lat_max_indo], cmap=ListedColormap([sea_color]))
img_indo = ax_indo.imshow(elevation_land_subset_indo, extent=[lon_min_indo, lon_max_indo, lat_min_indo, lat_max_indo], cmap=custom_cmap, norm=norm, alpha=0.9)
coastline.boundary.plot(ax=ax_indo, color='black', linewidth=0.5)
#ax_indo.set_title('Topography Map of Indonesia', fontsize=10)
ax_indo.set_xlabel('Longitude', fontsize=12)
ax_indo.set_ylabel('Latitude', fontsize=12)
ax_indo.set_xticks(np.arange(95, 145, 5))
ax_indo.set_yticks(np.arange(-10, 10, 5))
ax_indo.grid(True, linestyle="--", linewidth=0.7, alpha=0.7)

# **Draw Red Rectangle on Indo Map**
rect = patches.Rectangle(
    (lon_min_west, lat_min_west),  
    lon_max_west - lon_min_west,  
    lat_max_west - lat_min_west,  
    linewidth=2, edgecolor='red', facecolor='none', linestyle="--"
)
ax_indo.add_patch(rect)

# **Add Colorbar (Move to the Right)**
cbar_ax = fig.add_axes([0.77, 0.3, 0.02, 0.4])  # [left, bottom, width, height]
cbar = fig.colorbar(img_indo, cax=cbar_ax, ticks=elevation_bins, extend="both")
cbar.set_label('Elevation (meters)', fontsize=12)
cbar.ax.set_yticklabels([ 
    '< 300', '300 — 600', '600 — 900', '900 — 1200',
    '1200 — 1500', '1500 — 1800', '1800 — 2100', '> 2100'
], fontsize=10)
fig.subplots_adjust(left=0.05, right=0.95, bottom=0.1, top=0.9, wspace=0.15)
plt.show()
```

```python
# Create the figure
fig, ax_west = plt.subplots(figsize=(8, 5))  # Adjust figure size

# **Plot West Sumatra Topography**
ax_west.imshow(elevation_subset_west < 0, extent=[lon_min_west, lon_max_west, lat_min_west, lat_max_west], cmap=ListedColormap([sea_color]))
img_west = ax_west.imshow(elevation_land_subset_west, extent=[lon_min_west, lon_max_west, lat_min_west, lat_max_west], cmap=custom_cmap, norm=norm, alpha=0.9)

# **Plot Coastline and Stations**
coastline_west.boundary.plot(ax=ax_west, color='black', linewidth=1)
ax_west.scatter(locations['longitude'], locations['latitude'], color='blue', marker='o', edgecolors='w', linewidths=0.7, s=30, label='Stations', zorder=3)

# **Set Labels, Ticks, and Grid**
ax_west.set_xlabel('Longitude', fontsize=12)
ax_west.set_ylabel('Latitude', fontsize=12)
ax_west.set_xticks(np.arange(97, 103, 1))
ax_west.set_yticks(np.arange(-3, 2, 1))
ax_west.grid(True, linestyle="--", linewidth=0.7, alpha=0.7)

# **Legend**
ax_west.legend(fontsize=12)

# **Colorbar**
cbar = fig.colorbar(img_west, ax=ax_west, orientation='vertical', fraction=0.05, pad=0.02, ticks=elevation_bins, extend="both")
cbar.set_label('Elevation (meters)', fontsize=12)
cbar.ax.set_yticklabels([ 
    '< 300', '300 — 600', '600 — 900', '900 — 1200',
    '1200 — 1500', '1500 — 1800', '1800 — 2100', '> 2100'
], fontsize=10)

# **Show Plot**
plt.tight_layout()
plt.savefig('map.tiff', dpi=300, format='tiff')
plt.show()
```
